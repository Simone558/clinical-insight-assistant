```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
print("Setup ok ✓")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv",
                 header=None,
                 names=["Pregnancies","Glucose","BloodPressure","SkinThickness",
                        "Insulin","BMI","DiabetesPedigreeFunction","Age","Outcome"])
```

```python
df.shape
```

```python
df.head()
```

```python
df.dtypes
```

```python
col = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
print(col)
```

```python
for x in col:
    print('la colonna', x, 'ha', (df[x] == 0).sum(),'zeri')
```

```python
for x in col:
    df[x] = df[x].replace(0,np.nan)
```

```python
df.isnull().sum()
```

```python
y = df['Outcome']
print(y)
```

```python
x = df.drop('Outcome',axis = 1)
x.head()
```

```python
import seaborn as sns
sns.histplot(data=df,x='Glucose',hue='Outcome')

#l'outcome 1 (diabetico) è più frequente nelle osservazioni con valori alti di glucosio mentre i non diabetici sono concentrati nelle osservazioni 
#con valori bassi di glucosio
```

```python
col = df.drop('Outcome',axis=1)
for x in col:
    sns.histplot(data=df,x=x,hue='Outcome')
    plt.show()

#l'unica variabile che discrimina bene i diabetici dai non diabetici sembra essere 'Glucose'
```

```python
#class balance

print('la variabile Outcome presenta',(df['Outcome'] == 0).sum(),'valorizzazioni a 0')
print('la variabile Outcome presenta',(df['Outcome'] == 1).sum(),'valorizzazioni a 1')

df['Outcome'].value_counts()
# la variabile target è sbilanciata
```
